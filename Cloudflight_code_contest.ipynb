Level one task is to perform Principal Component Analysis
(PCA) on the data in the CSV file (find Level 1.pdf on this repo for more instructions)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# we will use sklearn for PCA
from sklearn.decomposition import PCA
```

```python
#so that we are able to access the two datasets from drive in colab
from google.colab import drive
drive.mount("/content/drive")
```

```python
data = pd.read_csv("/content/data.csv", header = None)
data.head()
```

```python
# we explore the data a little more before we perform the PCA
data.describe()
```

```python
data.info()
```

```python
# we now perform the PCA
X = data.values

# we decide to pick reduce the into two dimensions
pca = PCA(n_components=2)

# we call the PCA decomposition to fits the PCA model to the data
# Then, transform the original data into the new lower-dimensional space
X_new = pca.fit_transform(X)
X_new
```

```python
# the task of level one is to compute the standard deviation of the resulting two dimensions

# so we calculate the standard deviation along the first axis to get 1-dimensional array from each principle component
stds = X_new.std(axis=0)
print(stds)

# round off the standard deviation values to two decimal places
# and convert the result to a Python list.
output = list(np.around(stds, decimals=2))
output
```
